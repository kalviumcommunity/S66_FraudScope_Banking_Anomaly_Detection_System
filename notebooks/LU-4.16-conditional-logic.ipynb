# LU 4.16 - Conditional Statements for Data Logic

This notebook applies `if/elif/else` rules to classify transaction risk by amount.

```python
from src.fraudscope.conditional_rules import risk_band, summarize_risk

amounts = [1200, 5200, 15000, 4300, 8800]
labels = [(amount, risk_band(float(amount))) for amount in amounts]
summary = summarize_risk([float(value) for value in amounts])

print(labels)
print(summary)
```
